```python
import numpy as np

# Create an 8x8 array filled with zeros
checkerboard = np.zeros((8, 8), dtype=int)

# Set the alternating pattern for the checkerboard
checkerboard[1::2, ::2] = 1  # odd rows, even columns
checkerboard[::2, 1::2] = 1  # even rows, odd columns

# Print the checkerboard
print(checkerboard)
```

```python
import numpy as np

x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])
v = np.array([9,10])
w = np.array([11,12])

# 1. Add
addition = x + y
print("Addition:\n", addition)

# 2. Subtract
subtraction = x - y
print("Subtraction:\n", subtraction)

# 3. Multiplication
multiplication = x * 3
multiplication_1 = v * 2
print("Multiplication of x with 3:\n", multiplication)
print("Multiplication of v with 2:\n", multiplication_1)

# 4. Square
square_x = x ** 2
print("Square of x:\n", square_x)

# Dot product
VandW = np.dot(v, w)
print("Dot product between v and w: ", VandW)

XandV = np.dot(x, v)
print("Dot product between X and V: ", XandV)

XandY = np.dot(x, y)
print("Dot product between X and Y: \n", XandY)

# 6. Concatenation
XwithY = np.vstack((x, y))
print("Vertical stack of x and y:\n", XwithY)

VwithW = np.column_stack((v, w))
print("Column stack of v and w:\n", VwithW)

# 7. Fix concatenation error
xwithv = np.concatenate((x, v.reshape(1, -1)), axis=0)  # Reshape v to be a row vector
print("Concatenation of x and v:\n", xwithv)
```

```python
#problem4
import numpy as np

A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

I = np.eye(2, dtype=int)

A_inv = np.linalg.inv(A)
AdotAinv = np.dot(A,A_inv)
AdotAinv_int = np.round(AdotAinv).astype(int)
print(AdotAinv_int)

if np.array_equal(AdotAinv_int, I):
    print("True")
else:
    print("False")
```

```python
import numpy as np

A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

AdotB = np.dot(A,B)
BdotA = np.dot(B,A)

print(AdotB)
print(BdotA)

if np.array_equal(AdotB, BdotA):
    print("Both are equal")
else:
    print("Not equal")
```
