```python
def cost_function(X, Y, W):
    n = len(Y)
    Y_pred = np.dot(X, W)
    cost = (1 / (2 * n)) * np.sum((Y_pred - Y) ** 2)
    return cost
```

```python
X_test_case = np.array([[1, 2], [3, 4], [5, 6]])
Y_test_case = np.array([3, 7, 11])
W_test_case = np.array([1, 1])
print(cost_function(X_test_case, Y_test_case, W_test_case))
```

```python
def gradient_descent(X, Y, W, alpha, iterations):
    m = len(Y)
    cost_history = []

    for _ in range(iterations):
        Y_pred = np.dot(X, W)
        loss = Y_pred - Y
        dw = (1 / m) * np.dot(X.T, loss)
        W = W - alpha * dw
        cost_history.append(cost_function(X, Y, W))

    return W, cost_history
```

```python
np.random.seed(1)
X_rand = np.random.rand(100, 3)
Y_rand = np.random.rand(100)
W_rand = np.random.rand(3)
W_final, cost_hist = gradient_descent(X_rand, Y_rand, W_rand, 0.01, 1000)
print(W_final)
```

```python
def rmse(Y, Y_pred):
    return np.sqrt(np.mean((Y - Y_pred) ** 2))
```

```python
def r2(Y, Y_pred):
    mean_y = np.mean(Y)
    ss_tot = np.sum((Y - mean_y) ** 2)
    ss_res = np.sum((Y - Y_pred) ** 2)
    return 1 - (ss_res / ss_tot)
```

```python
W_init = np.zeros(X_train.shape[1])
alpha = 0.00001
iterations = 1000
W_opt, cost_hist = gradient_descent(X_train, Y_train, W_init, alpha, iterations)
Y_pred_test = np.dot(X_test, W_opt)
print("RMSE:", rmse(Y_test, Y_pred_test))
print("R2:", r2(Y_test, Y_pred_test))
```

```python
for lr in [0.000001, 0.00001, 0.001, 0.01]:
    W_lr, _ = gradient_descent(X_train, Y_train, W_init, lr, 500)
    Y_lr_pred = np.dot(X_test, W_lr)
    print("LR:", lr, "RMSE:", rmse(Y_test,Y_lr_pred))
```
