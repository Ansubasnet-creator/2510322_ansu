```python
# Step 2: Composite Development Score for South Asia (2022)

# Make a fresh copy of the subset
south_asia_df = hdi_south_asia.copy()

# Normalize life expectancy into an index
le_range = south_asia_df['life_expectancy'].max() - south_asia_df['life_expectancy'].min()
south_asia_df['life_exp_index'] = (
    (south_asia_df['life_expectancy'] - south_asia_df['life_expectancy'].min()) / le_range
)

# Normalize GNI per capita into an index
gni_range = south_asia_df['gross_inc_percap'].max() - south_asia_df['gross_inc_percap'].min()
south_asia_df['gni_index'] = (
    (south_asia_df['gross_inc_percap'] - south_asia_df['gross_inc_percap'].min()) / gni_range
)

# Build composite score (weighted average of indices)
south_asia_df['composite_score'] = (
    0.3 * south_asia_df['life_exp_index'] + 0.3 * south_asia_df['gni_index']
)

# Focus on 2022 data and rank countries
south_asia_2022 = south_asia_df.query("year == 2022")
ranked_scores = south_asia_2022.sort_values('composite_score', ascending=False)

# Show country and composite score
ranked_scores[['country', 'composite_score']]
```

```python
# Outlier Detection with IQR method
def find_outliers(data_series):
    # Compute quartiles
    q1_val, q3_val = data_series.quantile([0.25, 0.75])
    iqr_val = q3_val - q1_val

    # Define cutoff thresholds
    lower_cut = q1_val - 1.5 * iqr_val
    upper_cut = q3_val + 1.5 * iqr_val

    # Extract outlier values directly
    outliers = data_series[(data_series < lower_cut) | (data_series > upper_cut)]
    return outliers
```

```python
# Step 2.4: Outlier Detection for South Asia (2022)

# Define outlier detection function (IQR method)
def detect_outliers(series):
    q1, q3 = series.quantile([0.25, 0.75])
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return (series < lower) | (series > upper)

# Filter dataset for 2022
south_asia_2022 = hdi_south_asia.query("year == 2022").copy()

# Apply outlier detection
south_asia_2022['hdi_outlier'] = detect_outliers(south_asia_2022['hdi'])
south_asia_2022['gni_outlier'] = detect_outliers(south_asia_2022['gross_inc_percap'])

# Combine into a single flag
south_asia_2022['is_outlier'] = south_asia_2022['hdi_outlier'] | south_asia_2022['gni_outlier']

# Show results
south_asia_2022[['country', 'hdi', 'gross_inc_percap', 'is_outlier']]
```

```python
# Step 2.5: Scatter Plot for Outlier Detection (South Asia, 2022)

plt.figure(figsize=(9,6))

# Plot normal points
plt.scatter(
    south_asia_2022.loc[~south_asia_2022['is_outlier'], 'gross_inc_percap'],
    south_asia_2022.loc[~south_asia_2022['is_outlier'], 'hdi'],
    label='Normal',
    alpha=0.7,
    color='green',
    edgecolor='black'
)

# Plot outlier points
plt.scatter(
    south_asia_2022.loc[south_asia_2022['is_outlier'], 'gross_inc_percap'],
    south_asia_2022.loc[south_asia_2022['is_outlier'], 'hdi'],
    label='Outliers',
    alpha=0.9,
    color='red',
    marker='D'
)

plt.xlabel('Gross National Income per Capita', fontsize=12)
plt.ylabel('Human Development Index (HDI)', fontsize=12)
plt.title('South Asia (2022): HDI vs GNI Outlier Detection', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(axis='both', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()

# Insight note
print("Applying the 1.5  IQR rule to South Asian countries in 2022 shows no outliers. "
      "This is expected given the small sample size and relatively similar HDI and GNI values.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Filter South Asia data for 2022
hdi_south_asia_2022 = hdi_south_asia[hdi_south_asia['year'] == 2022]

# Step 2: Keep required columns and drop missing values
hdi_south_asia_2022 = hdi_south_asia_2022[['country', 'gross_inc_percap', 'hdi']].dropna()

# Step 3: Create gap variable
hdi_south_asia_2022['gni_hdi_gap'] = (
    hdi_south_asia_2022['gross_inc_percap'] - hdi_south_asia_2022['hdi']
)

# Step 4: Sort gaps
gap_desc = hdi_south_asia_2022.sort_values(by='gni_hdi_gap', ascending=False)
gap_asc = hdi_south_asia_2022.sort_values(by='gni_hdi_gap', ascending=True)

# Step 5: Select top 3 positive and negative gaps
top_positive = gap_desc.head(3)
top_negative = gap_asc.head(3)

# Step 6: Combine data
plot_data = pd.concat([top_positive, top_negative])

# Step 7: Plot
colors = ['green'] * len(top_positive) + ['red'] * len(top_negative)

plt.figure(figsize=(8, 5))
plt.barh(plot_data['country'], plot_data['gni_hdi_gap'], color=colors)
plt.axvline(0, color='black', linewidth=0.8)

plt.xlabel('GNIHDI Gap')
plt.ylabel('Country')
plt.title('Top 3 Positive and Negative GNIHDI Gaps (South Asia)')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```

```python
import pandas as pd

# Load the regional datasets
south_asia = pd.read_csv("HDI_SouthAsia_2020_2022.csv")
middle_east = pd.read_csv("HDI_MiddleEast_2020_2022.csv")

# Compute average HDI by country
sa_avg = south_asia.groupby('country')['hdi'].mean().reset_index()
me_avg = middle_east.groupby('country')['hdi'].mean().reset_index()

# Top 3 and Bottom 3 countries in South Asia
sa_top_3 = sa_avg.sort_values('hdi', ascending=False).head(3)
sa_bottom_3 = sa_avg.sort_values('hdi', ascending=True).head(3)

# Top 3 and Bottom 3 countries in Middle East
me_top_3 = me_avg.sort_values('hdi', ascending=False).head(3)
me_bottom_3 = me_avg.sort_values('hdi', ascending=True).head(3)

# Combine results into a single DataFrame
combined = pd.concat([
    sa_top_3.assign(region='South Asia', group='Top 3'),
    sa_bottom_3.assign(region='South Asia', group='Bottom 3'),
    me_top_3.assign(region='Middle East', group='Top 3'),
    me_bottom_3.assign(region='Middle East', group='Bottom 3')
])

combined.reset_index(drop=True, inplace=True)
combined
```

```python
import matplotlib.pyplot as plt

# Create labels with country and region
labels = combined.apply(lambda row: f"{row['country']} ({row['region']})", axis=1)

# Plot horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(labels, combined['hdi'], color='skyblue')
plt.axvline(combined['hdi'].mean(), color='red', linestyle='--', linewidth=1, label='Overall Mean HDI')

plt.xlabel('Average HDI (2020-2022)')
plt.ylabel('Country')
plt.title('Top and Bottom HDI Performers: South Asia vs Middle East')
plt.legend()
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# 5. HDI Disparity Analysis

# Compute HDI range for each region
sa_hdi_range = south_asia['hdi'].max() - south_asia['hdi'].min()
me_hdi_range = middle_east['hdi'].max() - middle_east['hdi'].min()

# Compute HDI mean and standard deviation
sa_hdi_mean = south_asia['hdi'].mean()
sa_hdi_std = south_asia['hdi'].std()
sa_hdi_cv = sa_hdi_std / sa_hdi_mean  # Coefficient of variation

me_hdi_mean = middle_east['hdi'].mean()
me_hdi_std = middle_east['hdi'].std()
me_hdi_cv = me_hdi_std / me_hdi_mean  # Coefficient of variation

# Combine results into a DataFrame
variation_stats = pd.DataFrame({
    'Region': ['South Asia', 'Middle East'],
    'HDI Range': [sa_hdi_range, me_hdi_range],
    'HDI Coefficient of Variation': [sa_hdi_cv, me_hdi_cv]
})

variation_stats
```

HDI Range: The difference between the highest and lowest HDI in the region.

Coefficient of Variation (CV): Relative variability of HDI; higher CV indicates more disparity.


```python
import numpy as np
import matplotlib.pyplot as plt

# 6. Correlation Analysis: HDI vs Gender Development and Life Expectancy

# Compute correlations
sa_corr_gender = south_asia['hdi'].corr(south_asia['gender_development'])
sa_corr_life = south_asia['hdi'].corr(south_asia['life_expectancy'])

me_corr_gender = middle_east['hdi'].corr(middle_east['gender_development'])
me_corr_life = middle_east['hdi'].corr(middle_east['life_expectancy'])

print(f"South Asia: HDI vs Gender Development Correlation = {sa_corr_gender:.2f}")
print(f"South Asia: HDI vs Life Expectancy Correlation = {sa_corr_life:.2f}")
print(f"Middle East: HDI vs Gender Development Correlation = {me_corr_gender:.2f}")
print(f"Middle East: HDI vs Life Expectancy Correlation = {me_corr_life:.2f}")

# Scatter plot with regression line for South Asia: HDI vs Gender Development
x = south_asia['gender_development']
y = south_asia['hdi']

# Remove missing values
mask = x.notna() & y.notna()
x, y = x[mask], y[mask]

# Fit linear regression
slope, intercept = np.polyfit(x, y, 1)

# Plot
plt.figure(figsize=(7,5))
plt.scatter(x, y, color='skyblue', edgecolor='k', alpha=0.7, label='Countries')
plt.plot(x, slope*x + intercept, color='red', linewidth=2, label=f'Fit: y={slope:.2f}x+{intercept:.2f}')
plt.xlabel('Gender Development Index')
plt.ylabel('HDI')
plt.title('South Asia: HDI vs Gender Development')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
