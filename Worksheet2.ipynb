```python
medical_df = pd.read_csv(
    '/content/drive/MyDrive/medical_students_dataset.csv'
)
medical_df.head()
```

```python
medical_df.info()
medical_df.isnull().sum()
```

```python
# Numerical columns  Mean Imputation
num_cols = medical_df.select_dtypes(include=np.number).columns
medical_df[num_cols] = medical_df[num_cols].fillna(medical_df[num_cols].mean())

# Categorical columns  Mode Imputation
cat_cols = medical_df.select_dtypes(include='object').columns
for col in cat_cols:
    medical_df[col].fillna(medical_df[col].mode()[0], inplace=True)
```

```python
medical_df.duplicated().sum()
medical_df = medical_df.drop_duplicates()
```

```python
titanic_df = pd.read_csv(
    '/content/drive/MyDrive/Titanic-Dataset.csv'
)
titanic_df.head()
```

```python
subset_df = titanic_df[['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']]
first_class_df = subset_df[subset_df['Pclass'] == 1]
first_class_df.head()
```

```python
first_class_df['Fare'].agg(['mean', 'median', 'max', 'min'])
```

```python
first_class_df['Age'].isnull().sum()
first_class_df = first_class_df.dropna(subset=['Age'])
```

```python
embarked_dummies = pd.get_dummies(
    titanic_df['Embarked'],
    prefix='Embarked'
)

titanic_df = pd.concat([titanic_df, embarked_dummies], axis=1)
titanic_df.drop('Embarked', axis=1, inplace=True)

titanic_df.head()
```

```python
titanic_df.groupby('Sex')['Survived'].mean()
```

```python
sns.barplot(x='Sex', y='Survived', data=titanic_df)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
sns.catplot(
    x='Sex',
    y='Survived',
    hue='Embarked_C',
    data=titanic_df,
    kind='bar'
)
plt.show()
```

```python
titanic_df['AgeGroup'] = pd.qcut(titanic_df['Age'], 5)
```

```python
survival_stats = titanic_df.groupby(
    ['Pclass', 'AgeGroup']
)['Survived'].mean()

survival_stats
```

```python
survival_stats.unstack().plot(kind='bar', figsize=(10,6))
plt.title('Survival Rate by Age Group and Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```
